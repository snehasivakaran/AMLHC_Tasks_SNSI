```python
from sklearn import datasets
import pandas as pd

# Load the iris dataset as a pandas dataframe
iris_X, iris_y = datasets.load_iris(return_X_y=True, as_frame=True)

# Display the first rows
iris_X.head()
```

```python
# Function to calculate mean values for all columns of a dataframe
def calcMeans(df):
    means = df.apply(lambda column: column.mean())
    return means
```

```python
# Calculate the mean values of the iris features
mean_values = calcMeans(iris_X)

mean_values
```
