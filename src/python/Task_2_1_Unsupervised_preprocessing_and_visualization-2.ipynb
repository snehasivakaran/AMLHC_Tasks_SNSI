```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# 1. Load dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/food.csv', index_col=0)
```

```python
# Analyze dataset
print("--- Shape ---")
print(df.shape)

print("\n--- Missing Values ---")
print(df.isnull().sum())

print("\n--- Feature Statistics ---")
print(df.describe())
```

```python
# 3. Preprocess data using z-transformation
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
# 4. Perform PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
```

```python
# 5. Create score plot
plt.figure(figsize=(10, 8))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c='blue', edgecolors='k')

# Label axes
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Score Plot - Food Dataset")

# Add country labels
for i, txt in enumerate(df.index):
    plt.annotate(txt, (pca_data[i, 0], pca_data[i, 1]), xytext=(5,0), textcoords='offset points')

plt.grid(True)
plt.show()
```
